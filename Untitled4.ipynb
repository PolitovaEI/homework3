```python
import matplotlib.pyplot as plt
import numpy as np
pplt=np.loadtxt("http://www.scipy-lectures.org/_downloads/populations.txt", delimiter = "\t")


#No1

plt.plot(pplt[:,0],pplt[:,1], label='hares')
plt.plot(pplt[:,0],pplt[:,2], label='lynxes')
plt.plot(pplt[:,0],pplt[:,3], label='carrot')
plt.xticks(np.arange(pplt[:,0].min(),pplt[:,0].max(),2))
plt.axvline(pplt[np.argmax(pplt[:,1]),0])
plt.axvline(pplt[np.argmax(pplt[:,2]),0], color='orange')
plt.axvline(pplt[np.argmax(pplt[:,3]),0], color='green')
plt.legend()
plt.show()
```

```python
#No2

total=pplt[:,1:3].sum(axis=1)
plt.bar(pplt[:,0], pplt[:,1]/total,label="hares")
plt.bar(pplt[:,0], pplt[:,2]/total,bottom=pplt[:,1]/total,label="lynxes")
plt.legend()
plt.show()
```

```python
#No3

fig, axs = plt.subplots(1, 3, figsize=(15, 3))
axs[0].hist(pplt[:,1],bins = 10)
axs[0].axvline(pplt[:,1].mean())
axs[1].hist(pplt[:,2],bins = 10)
axs[1].axvline(pplt[:,2].mean())
axs[2].hist(pplt[:,3],bins = 10)
axs[2].axvline(pplt[:,3].mean())
plt.show()

mean = np.mean(pplt[:, 1:4], axis = 0)
std = np.std(pplt[:, 1:4], axis = 0)
print("Mean: ")
print(mean)
#print("Std: ")
print(std)
```

```python
#No4



maxima = pplt[pplt.argmax(axis = 0), 0]
minima = pplt[pplt.argmin(axis = 0), 0]

plt.subplot(1, 3, 1)
plt.plot(pplt[:, 0], pplt[:, 1])
plt.ylabel('Hares')
plt.xlabel('Years')
plt.axvline(x=maxima[1], color='r', linestyle='dashed')
plt.axvline(x=minima[1], color='y', linestyle='dashed')

plt.subplot(1, 3, 2)
plt.plot(pplt[:, 0], pplt[:, 2])
plt.xlabel('Years')
plt.ylabel('Lynxes')
plt.axvline(x=maxima[2], color='r', linestyle='dashed')
plt.axvline(x=minima[2], color='y', linestyle='dashed')

plt.subplot(1, 3, 3)
plt.plot(pplt[:, 0], pplt[:, 3])
plt.xlabel('Years')
plt.ylabel('Carrots')
plt.axvline(x=maxima[3], color='r', linestyle='dashed')
plt.axvline(x=minima[3], color='y', linestyle='dashed')

plt.show()
```

```python
#No5


mxm = np.argmax(np.vstack((np.array(pplt[:,1]), np.array(pplt[:,2]))),axis = 0)
animals = np.empty(mxm.size, dtype = np.str)
mask1 = mxm==1
mask2 = mxm==0
animals[mask1] = 'L'
animals[mask2] = 'H'
print(animals)
label = 'hares','lynxes'
sizes = [mxm.shape[0],mxm.sum()]
plt.pie(sizes, startangle=90,labels = label)
plt.show()
```

```python
#No6


hrs = np.array(pplt[np.argwhere(pplt[:,1] > 50000),0]);
lnx = np.array(pplt[np.argwhere(pplt[:,2] > 50000),0])
plt.plot(pplt[:,0],pplt[:,1], label='hares') 
plt.plot(pplt[:,0],pplt[:,2], label='lynxes')
plt.plot(pplt[:,0],pplt[:,3], label='carrots')
plt.xticks(np.arange(pplt[:,0].min(),pplt[:,0].max(),2))
for year in hrs:
    plt.axvline(year, color='red')
for year in lnx:
    plt.axvline(year, color='black')
plt.legend()
plt.show()
```

```python
#No7


minpop = np.transpose(np.argsort(pplt[:,1:], axis = 0)[:2])
print(pplt[minpop,0])
```

```python
#No8



hares1 = np.delete(pplt[:, 1], -1)
hares2 = np.delete(pplt[:, 1], 0)

difference = hares2 - hares1
newtime = np.delete(pplt[:,0], 0)

plt.plot(newtime, difference,label="Difference")
plt.plot(pplt[:,0], pplt[:,2], label = "Lynxes")
plt.xticks(np.arange(pplt[:,0].min(),pplt[:,0].max(),2))
print("Correlation coeffitient = ", np.corrcoef(difference, np.delete(pplt[:,2],0))[0,1])
plt.legend()
plt.show()
```
